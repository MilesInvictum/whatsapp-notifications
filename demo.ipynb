```python
def get_forecast(response, day_offset:"int"=0, day_hour:"int"=12):
    """
    Extracts data from the JSON response of weather api

    :param response: :class:`response` object.
    :param day_offset: for predicted day.
    :param day_hour: for the hour of the day on a 24 hours format.
    :rtype: Any
    """
    date = response['forecast']['forecastday'][day_offset]['date']
    hour = int(response['forecast']['forecastday'][day_offset]['hour'][day_hour]['time'].split()[1].split(':')[0])
    w_condition = response['forecast']['forecastday'][day_offset]['hour'][day_hour]['condition']['text']
    temp_c = response['forecast']['forecastday'][day_offset]['hour'][day_hour]['temp_c']
    humidity = response['forecast']['forecastday'][day_offset]['hour'][day_hour]['humidity']
    rain = response['forecast']['forecastday'][day_offset]['hour'][day_hour]['will_it_rain']
    rain_ch = response['forecast']['forecastday'][day_offset]['hour'][day_hour]['chance_of_rain']

    maxtemp_c = response['forecast']['forecastday'][day_offset]['day']['maxtemp_c']
    mintemp_c = response['forecast']['forecastday'][day_offset]['day']['mintemp_c']
    avgtemp_c = response['forecast']['forecastday'][day_offset]['day']['avgtemp_c']

    return date, hour, w_condition, temp_c, humidity, rain, rain_ch, maxtemp_c, mintemp_c, avgtemp_c
```

```python
data = []
for i in tqdm(range(len(response['forecast']['forecastday'][0]['hour'])), colour='blue'):
    data.append(get_forecast(response,0,i))
```

```python
col = ['Date', 'Hour', 'Condition', 'Temp C', 'Humidity', 'Rain', 'Rain Chance', 'Max Temp C', 'Min Temp C', 'Avg Temp C'] # create columns names for data frame
df = pd.DataFrame(data=data,columns=col) # Create data frame to filter data to send
df = df.sort_values(by = 'Hour', ascending=True)
df
```

```python
df_rain = df[(df['Rain'] == 0) & (df['Hour'] > 6) & (df['Hour'] < 22)] # filter data (all columns in place)
df_rain = df_rain[['Hour','Condition','Temp C','Humidity','Rain Chance']] # select columns of interest
df_rain.set_index('Hour', inplace=True)
df_rain
```
